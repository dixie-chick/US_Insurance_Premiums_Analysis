```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance_cleaned.csv')
df
```

```python
bmi_charges = sns.scatterplot(data=df, x="bmi", y="charges")
plt.xlabel('BMI')
plt.ylabel('Charges ($)')
plt.title('BMI vs. Charges')
bmi_charges.get_figure().savefig('images/bmi_vs_charges.png')
```

```python
age_charges = sns.scatterplot(data=df, x="age", y="charges")
plt.xlabel('Age')
plt.ylabel('Charges ($)')
plt.title('Age vs. Charges')
age_charges.get_figure().savefig('images/age_vs_charges.png')
```

```python
num_child_charges = sns.boxplot(data=df, x="children", y="charges")
plt.xlabel('Number of Children')
plt.ylabel('Charges ($)')
plt.title('# of Children vs. Charges')
num_child_charges.get_figure().savefig('images/num_children_vs_charges.png')
```

```python
smoker_charges = sns.boxplot(data=df, x="smoker_yes", y="charges")
plt.xlabel('Smoker Status (0 = no, 1 = yes)')
plt.ylabel('Charges ($)')
plt.title('Smoker Status vs. Charges')
smoker_charges.get_figure().savefig('images/smoker_vs_charges.png')
```

```python
sex_charges = sns.boxplot(data=df, x="sex_female", y="charges")
plt.xlabel('Sex (0 = Male, 1 = Female)')
plt.ylabel('Charges ($)')
plt.title('Sex vs. Charges')
sex_charges.get_figure().savefig('images/sex_vs_charges.png')
```

```python
corrs = df.corr(method='pearson').drop(['charges', 'smoker_no', 'sex_male'], axis=0)['charges']
corr_plot = sns.barplot(x=corrs.index.to_list(), y=np.abs(corrs))
plt.xticks(rotation=90)
plt.xlabel('Variable Name')
plt.ylabel('Correlation vs. charges variable')
plt.title('Correlations')
corr_plot.get_figure().savefig('images/corr_plot.png')
```
